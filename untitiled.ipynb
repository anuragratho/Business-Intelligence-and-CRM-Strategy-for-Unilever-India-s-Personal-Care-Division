```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python

# Load the dataset
df = pd.read_csv("dataset.csv")
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
personal_care_df = df[df['Product_Category'] == 'Personal Care']
```

```python
sns.set(style="whitegrid")
```

```python
plt.figure(figsize=(18, 20))
```

```python
# 1. Sales Volume Over Time
plt.figure(figsize=(12, 6))
sales_by_date = personal_care_df.groupby('Date')['Sales_Volume'].sum()
sales_by_date.plot(color='teal')
plt.title("Sales Volume Trend Over Time (Personal Care)")
plt.xlabel("Date")
plt.ylabel("Total Sales Volume")
plt.tight_layout()
plt.show()
```

```python
# 2. Sales Volume by Store Location
plt.figure(figsize=(10, 6))
sns.boxplot(data=personal_care_df, x='Store_Location', y='Sales_Volume', palette='pastel')
plt.title("Sales Volume Distribution by Store Location")
plt.xlabel("Store Location")
plt.ylabel("Sales Volume")
plt.tight_layout()
plt.show()
```

```python
# 3. Sales Volume Distribution by Promotion (Violin Plot)
plt.figure(figsize=(10, 6))
sns.violinplot(data=personal_care_df, x='Promotion', y='Sales_Volume', palette='Set2')
plt.title("Sales Volume Distribution by Promotion Status")
plt.xlabel("Promotion (0 = No, 1 = Yes)")
plt.ylabel("Sales Volume")
plt.tight_layout()
plt.show()
```

```python
# 4. Replenishment Lead Time vs Stock Level (Scatter Plot)
plt.figure(figsize=(10, 6))
sns.scatterplot(data=personal_care_df, x='Replenishment_Lead_Time', y='Stock_Level', hue='Store_Location', palette='deep')
plt.title("Replenishment Lead Time vs Stock Level")
plt.xlabel("Lead Time (Days)")
plt.ylabel("Stock Level")
plt.tight_layout()
plt.show()
```
